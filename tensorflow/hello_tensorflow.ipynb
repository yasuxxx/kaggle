```python
# mnist手写数据集的识别
import tensorflow as tf
(x_train,y_train),(x_test,y_test) = tf.keras.datasets.mnist.load_data()
# 将像素归一化 （0～255）->（0～1）
x_train,x_test = x_train/255.0,x_test/255.0

# 网络层数设置
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),  # 将28*28的图片摊开成一维输入
    tf.keras.layers.Dense(128,activation='relu'),  # 设置第一层网络，128个节点，激活函数采用relu
    tf.keras.layers.Dropout(0.2),                  # 设置20%的节点失活，防止模型overfit
    tf.keras.layers.Dense(10,activation='softmax'),# 输出层设置为10个节点（数字0～9），激活函数采用多分类的softmax
])

# 优化方式选择adam  损失函数选择多分类的sparse_categorical_crossentropy  评测选用accuracy
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

# 模型训练
model.fit(x_train,y_train,epochs=5)

# 评估模型效果
model.evaluate(x_test,y_test)
```
